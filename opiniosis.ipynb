Punctuation removal

```
import string
string.punctuation
```

```
#join function is used to recreate the sentences after removing the punctuations

def remove_punct(text):
  text_nopunctuation="".join([char for char in text if char not in string.punctuation])
  return text_nopunctuation
```

```
#build a lambda function to implement the above punctuation removal code
df1['text_with_nopunctuation'] = df1['text'].apply(lambda x: remove_punct(x))
```

```
# three columns 1 is the type of data 2nd is the normal text as per the dataset 3rd is the text without punctuation
df1.head() 
```

Tokenization

```
import re

def tokenize(text):
  tokens = re.split('\W+',text)
  return tokens
```

```
df1['text_with_tokens'] = df1['text_with_nopunctuation'].apply(lambda x: tokenize(x.lower()))
```

```
# introduced another column with tokenized values from the text without punctuation column
df1.head()
```

Removal of stopwords

```
import nltk
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
```

```
print(stopword)
```

```
def remove_stopwords(text_with_tokens):
  text_new = [word for word in text_with_tokens if word not in stopword]
  return text_new
```

```
df1['text_without_stopwords'] = df1['text_with_tokens'].apply(lambda x: remove_stopwords(x))
```

```
#introduced a new column without stopwords
df1.head()
```

Applying Porter Stemmer

```
import nltk
ps = nltk.PorterStemmer()
```

```
dir(ps)
```

```
def stemming(tokenized_text):
  text3 = [ps.stem(word) for word in tokenized_text]
  return text3
```

```
df1['text_stem'] = df1['text_without_stopwords'].apply(lambda x: stemming(x))
```

```
# new column for stemmed text
df1.head()
```

Applying lemmatization

```
 import nltk
 nltk.download('wordnet')
```

```
import nltk
from nltk.stem import WordNetLemmatizer 

# Init the Wordnet Lemmatizer
lemmatizer = WordNetLemmatizer()
```

```
dir(lemmatizer)
```

```
def lemmatizing(tokenized_text):
  text4 = [lemmatizer.lemmatize(word) for word in tokenized_text]
  return text4
  
```

```
df1['text_lemma'] = df1['text_without_stopwords'].apply(lambda x: lemmatizing(x))
```
